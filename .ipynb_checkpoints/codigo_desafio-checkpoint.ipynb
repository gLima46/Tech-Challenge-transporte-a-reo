```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score, precision_recall_fscore_support, confusion_matrix
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
```

```python
df_flights = pd.read_csv('dados/flights.csv', low_memory=False).sample(n=100000, random_state=42)
df_airlines = pd.read_csv('dados/airlines.csv')
df_airports = pd.read_csv('dados/airports.csv')
df_clima = pd.read_csv('dados/clima_aeroportos_2015.csv')
```

```python
df_flights['DATA_VOO'] = pd.to_datetime(df_flights[['YEAR', 'MONTH', 'DAY']].rename(columns={
    'YEAR': 'year', 
    'MONTH': 'month', 
    'DAY': 'day'
}))
```

```python
df_clima['DATA'] = pd.to_datetime(df_clima['DATA'])
```

```python
df = df_flights.merge(
    df_airlines, 
    left_on='AIRLINE', 
    right_on='IATA_CODE', 
    how='left'
).rename(columns={'AIRLINE_y': 'AIRLINE_NAME'}).drop('IATA_CODE', axis=1)
```

```python
df = df.merge(
    df_airports,
    left_on='DESTINATION_AIRPORT',
    right_on='IATA_CODE',
    how='left',
    suffixes=('', '_DEST') 
).drop('IATA_CODE', axis=1).rename(columns={
    'AIRPORT': 'DEST_AIRPORT_NAME',
    'CITY': 'DEST_CITY',
    'STATE': 'DEST_STATE',
    'LATITUDE': 'DEST_LATITUDE',
    'LONGITUDE': 'DEST_LONGITUDE'
})
```

```python
df = df.merge(
    df_clima,
    left_on=['DATA_VOO', 'ORIGIN_AIRPORT'],
    right_on=['DATA', 'IATA_CODE'],
    how='left'
).drop(['DATA', 'IATA_CODE'], axis=1)
```

```python
df = df.merge(
    df_clima,
    left_on=['DATA_VOO', 'DESTINATION_AIRPORT'],
    right_on=['DATA', 'IATA_CODE'],
    how='left',
    suffixes=('_ORIGIN', '_DEST') 
).drop(['DATA', 'IATA_CODE'], axis=1)
```

```python
df.head()
```

```python
df["IS_CANCELED"] = df["CANCELLED"].apply(lambda x: 1 if x == 1 else 0)
df["IS_DELAYED"] = df["ARRIVAL_DELAY"].apply(lambda x: 1 if x > 15 else 0)
```

```python
df.head()
```

```python
df.fillna(0, inplace=True)
```

```python
df.head()
```

```python
novos_nomes = {
    'tavg_ORIGIN':'TEMP_MEDIA_ORIGEM',
    'prcp_ORIGIN':'PRECIPITACAO_ORIGEM',
    'snow_ORIGIN':'NEVE_ORIGEM',
    'tmin_ORIGIN':'TEMP_MIN_ORIGEM',
    'tmax_ORIGIN':'TEMP_MAX_ORIGEM',
    'tavg_DEST':'TEMP_MEDIA_DESTINO',
    'prcp_DEST':'PRECIPITACAO_DESTINO',
    'snow_DEST':'NEVE_DESTINO',
    'tmin_DEST':'TEMP_MIN_DESTINO',
    'tmax_DEST':'TEMP_MAX_DESTINO'
}

df = df.rename(columns=novos_nomes)
```

```python
df.head()
```

```python
colunas_interesse = [
    'DEPARTURE_DELAY',  
    'ARRIVAL_DELAY',    
    'DISTANCE',         
    'TEMP_MEDIA_ORIGEM', 
    'PRECIPITACAO_ORIGEM',       
    'NEVE_ORIGEM',       
    'TEMP_MEDIA_DESTINO',
    'PRECIPITACAO_DESTINO',
    'NEVE_DESTINO'
]

cols_existentes = [c for c in colunas_interesse if c in df.columns]

matriz_correlacao = df[cols_existentes].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(
    matriz_correlacao, 
    annot=True,          
    fmt=".2f",           
    cmap='coolwarm',     
    center=0,            
    linewidths=0.5
)
plt.title('Correlação: Atrasos de Voo vs. Clima (Origem e Destino)')
plt.show()
```

```python
features = [
    'MONTH', 'DAY', 'DAY_OF_WEEK', 'SCHEDULED_DEPARTURE', 
    'SCHEDULED_TIME', 'DISTANCE', 'AIRLINE_x','ORIGIN_AIRPORT', 
    'DESTINATION_AIRPORT','TEMP_MEDIA_ORIGEM', 'PRECIPITACAO_ORIGEM', 
    'NEVE_ORIGEM','TEMP_MEDIA_DESTINO', 'PRECIPITACAO_DESTINO', 'NEVE_DESTINO'
]

target = 'IS_DELAYED'

X = df[features].copy()
y = df[target].copy()

top_airlines = X['AIRLINE_x'].value_counts().nlargest(15).index
X = X[X['AIRLINE_x'].isin(top_airlines)]
y = y[X.index] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

numeric_features = [
    'MONTH', 'DAY', 'SCHEDULED_DEPARTURE', 'SCHEDULED_TIME', 'DISTANCE',
    'TEMP_MEDIA_ORIGEM', 'PRECIPITACAO_ORIGEM', 'NEVE_ORIGEM',
    'TEMP_MEDIA_DESTINO', 'PRECIPITACAO_DESTINO', 'NEVE_DESTINO'
]

categorical_features = ['DAY_OF_WEEK', 'AIRLINE_x', 'ORIGIN_AIRPORT', 'DESTINATION_AIRPORT']

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')), 
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

```python
print("------------------- Regressão Logística -------------------")
lr_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression(random_state=42, max_iter=1000))
])
lr_pipeline.fit(X_train, y_train)
y_pred_lr = lr_pipeline.predict(X_test)

print("Acurácia:", accuracy_score(y_test, y_pred_lr))
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred_lr, target_names=['Não Atrasado', 'Atrasado']))

plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_lr), annot=True, fmt='d', cmap='Blues')
plt.title("Matriz de Confusão - Regressão Logística")
plt.xlabel("Previsto")
plt.ylabel("Real")
plt.show()
```

```python
print("------------------- Árvore de Decisão -------------------")
dt_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', DecisionTreeClassifier(random_state=42, max_depth=10))
])
dt_pipeline.fit(X_train, y_train)
y_pred_dt = dt_pipeline.predict(X_test)

print("Acurácia:", accuracy_score(y_test, y_pred_dt))
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred_dt, target_names=['Não Atrasado', 'Atrasado']))

plt.figure(figsize=(5, 4))
sns.heatmap(confusion_matrix(y_test, y_pred_dt), annot=True, fmt='d', cmap='Greens')
plt.title("Matriz de Confusão - Árvore de Decisão")
plt.xlabel("Previsto")
plt.ylabel("Real")
plt.show()
```

```python
print("------------------- K-Means -------------------")

airline_stats = df.groupby('AIRLINE_NAME')[['DEPARTURE_DELAY', 'ARRIVAL_DELAY', 'DISTANCE']].mean()

airline_stats = airline_stats.dropna()

scaler_cluster = StandardScaler()
X_cluster = scaler_cluster.fit_transform(airline_stats)


kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
airline_stats['CLUSTER'] = kmeans.fit_predict(X_cluster)


plt.figure(figsize=(12, 8))
sns.scatterplot(
    data=airline_stats, 
    x='DEPARTURE_DELAY', 
    y='ARRIVAL_DELAY', 
    hue='CLUSTER', 
    size='DISTANCE', 
    sizes=(50, 400), 
    palette='viridis',
    style='CLUSTER'
)

for i in range(airline_stats.shape[0]):
    plt.text(
        x=airline_stats.iloc[i]['DEPARTURE_DELAY'] + 0.2,
        y=airline_stats.iloc[i]['ARRIVAL_DELAY'], 
        s=airline_stats.index[i], 
        fontsize=9,
        alpha=0.7
    )

plt.title('Clusterização de Cias Aéreas: Atraso Partida vs. Chegada (Tamanho = Distância Média)')
plt.xlabel('Média Atraso Partida (min)')
plt.ylabel('Média Atraso Chegada (min)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Cluster ID', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("\nPerfil Médio de Cada Cluster:")
print(airline_stats.groupby('CLUSTER')[['DEPARTURE_DELAY', 'ARRIVAL_DELAY', 'DISTANCE']].mean())
```

```python
aeroportos_stats = df.groupby('AIRLINE_NAME').agg(
    total_voos=('IS_DELAYED', 'count'),
    media_atraso=('DEPARTURE_DELAY', 'mean'),
    perc_atrasos=('IS_DELAYED', 'mean') 
).reset_index()

top_criticos = aeroportos_stats[aeroportos_stats['total_voos'] > 50].sort_values('media_atraso', ascending=False).head(10)

print("Top 10 Companhias Mais Críticos (Maior Média de Atraso na Partida):")
print(top_criticos[['AIRLINE_NAME', 'media_atraso', 'perc_atrasos']])

plt.figure(figsize=(10, 6))
sns.barplot(data=top_criticos, x='media_atraso', y='AIRLINE_NAME', palette='magma')
plt.title('Top 10 Companhias com Maior Média de Atraso (min)')
plt.xlabel('Média de Atraso (min)')
plt.ylabel('')
plt.show()
```

```python
df['TEM_NEVE'] = np.where(df['NEVE_ORIGEM'] > 0, 'Sim', 'Não')
coluna_neve = 'NEVE_ORIGEM'

sns.barplot(
    data=df, 
    x='TEM_NEVE', 
    y='IS_DELAYED', 
    palette='coolwarm'
)

plt.title('Probabilidade de Atraso: Com Neve vs Sem Neve')
plt.ylabel('Probabilidade de Atraso (Média de IS_DELAYED)')
plt.show()
```

```python
df['HORA_PARTIDA'] = (df['SCHEDULED_DEPARTURE'] // 100).astype(int)

heatmap_data = df.pivot_table(
    index='HORA_PARTIDA', 
    columns='DAY_OF_WEEK', 
    values='DEPARTURE_DELAY', 
    aggfunc='mean'
)

plt.figure(figsize=(10, 10))
sns.heatmap(heatmap_data, cmap='RdYlGn_r', annot=False, fmt=".1f") 

plt.title('Mapa de Calor de Atrasos: Hora do Dia vs Dia da Semana')
plt.ylabel('Hora do Dia (0-23)')
plt.xlabel('Dia da Semana (1=Seg, 7=Dom)')
plt.show()
    
```

```python
df_airports_cluster = df.groupby('ORIGIN_AIRPORT').agg({
    'DEPARTURE_DELAY': 'mean',  
    'DISTANCE': 'mean',         
    'IS_DELAYED': 'count'       
}).rename(columns={'IS_DELAYED': 'VOLUME_VOOS'})

df_airports_cluster = df_airports_cluster[df_airports_cluster['VOLUME_VOOS'] > 50]

scaler = StandardScaler()
X_air = scaler.fit_transform(df_airports_cluster)

kmeans_air = KMeans(n_clusters=4, random_state=42, n_init=10)
df_airports_cluster['CLUSTER'] = kmeans_air.fit_predict(X_air)

plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df_airports_cluster,
    x='DEPARTURE_DELAY',
    y='VOLUME_VOOS', 
    hue='CLUSTER',
    palette='deep',
    s=100
)

plt.title('Clusterização de Aeroportos: Volume vs. Atraso Médio')
plt.xlabel('Média de Atraso (min)')
plt.ylabel('Volume de Voos (Quantidade)')
plt.grid(True, alpha=0.3)
plt.show()

print(df_airports_cluster.groupby('CLUSTER').mean())
```
